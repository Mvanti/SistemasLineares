# Análise de Sistemas em Tempo Contínuo com a Transformada de Laplace 

## Função de Transferência e Teorema da Convolução

Anteriormente vimos que para um sistema de estado nulo a equação entrada-saída pode ser escrita como

$$Q(s)Y(s)=P(s)X(s)$$

onde $X(s)=\mathcal{L}(x(t))$ e $Y(s)=\mathcal{L}(y(t))$.  A função racional

$$H(s)=\frac{Y(s)}{X(s)}$$

é a função de transferência do sistema. Podemos representar a relação entrada/saída no sistema no domínio do tempo ou no domínio da frequência complexa nas forma mostradas nas figuras abaixo. Desde que 
\begin{equation*}
y(t)=x(t)*h(t),
\label{}\tag{1}
\end{equation*}
e da definição de função de transferência,
\begin{equation*}
Y(s)=H(s)X(s),
\label{}\tag{2}
\end{equation*}

podemos concluir, de (1) e (2) que 
\begin{equation*}
\mathcal{L}(x(t)*h(t))=H(s)X(s)
\label{}\tag{3}
\end{equation*}
Este importante resultado, que pode ser generalizado para qualquer par de funções com Transformada de Laplace, é o *Teorema da Convolução*.
Considere agora que a entrada é um impulso, isto é,  $x(t)=\delta(t)$, daí 

$$y(t)= h(t)*\delta(t)=\int_{o} ^{\infty} h(\tau)\delta(t-\tau)d\tau=\int_{o} ^{\infty} h(\tau)\delta(\tau-t)d\tau =h(t)$$
devido a propriedade da amostragem do impulso. Com efeito, a convolução de uma função com o impulso é igual a própria função. Finalmente, como $X(s)=1$, de (2) temos  $Y(s)=H(s)$, e portanto 

$$\mathcal{L}(h(t))=H(s)$$

*A função de transferência é a transformada de laplace da Resposta ao impulso.*

## Teoremas dos Valores Inicial e Final

### Teorema do Valor Inicial

Seja um sinal $x(t)$. A transformada de Laplace de sua derivada é

$$\mathcal{L}(x'(t)u(t))=sX(s)-x(0^-)=\int_{0^-}^{\infty} x'(t)e^{-st}\,dt=\int_{0^-}^{0^+}x'(t)e^{-st}\,dt+\int_{0^+}^{\infty} x'(t)e^{-st}\,dt,$$

$$sX(s)-x(0^-)=x(t)\big|_{0^-}^{0^+}+\int_{0^+}^{\infty} x'(t)e^{-st}\,dt=x(0^+)-x(0^-)+\int_{0^+}^{\infty} x'(t)e^{-st}\,dt\quad\mbox{e}$$

$$sX(s)=x(0^+)+\int_{0^+}^{\infty} x'(t)e^{-st}\,dt.$$

Tomando o limite $s\rightarrow \infty$ na equação acima (desde que o limite exista), temos

$$\lim_{s\to\infty} sX(s)=\lim_{s\to\infty}\int_{0^+}^{\infty} x'(t)e^{-st}\,dt+x(0^+).$$

Como o limite da integral na equação acima é nulo, o Teorema do Valor Inicial afirma que 

$$x(0^+)=\lim_{s\to\infty}sX(s)$$

**Observação Importante** Se $X(s)$ não é uma função racional estritamente própria $(N>M)$, então o limite de $sX(s)$ não existe e o teorema não pode ser aplicado diretamente, como mostra o exemplo a seguir: considere a expressão

$$Y(s)=\frac{s^2+1}{s^2+2s+1}$$

Claramente, $\lim_{s\to\infty}sY(s)=\infty$. Por outro lado, podemos fazer uma divisâo longa para escrever $Y(s)$ em termos de uma função estritamente própria, isto é,
$$Y(s)=1-\frac{2s}{s^2+2s+1}=1-\frac{2s}{(s+1)^2}=1+\frac{2}{(s+1)^2}-\frac{2}{s+1}$$

cuja tranformada inversa é

$$y(t)=\delta(t)+2te^{-t}-2e^{-t}$$

em $t=0^+$ o impulso desaparece, portanto o valor inicial $y(0^+)$ pode ser calculado como

$$y(0^+)=\lim_{s\to\infty}s\frac{-2s}{s^2+2s+1}=\lim_{s\to\infty}\frac{-2s^2}{s^2+2s+1}=-2.$$

### Teorema do Valor Final

Tomamos agora o limite quando $s\rightarrow 0$ na transformada de Laplace da derivada de $x(t)$, isto é

$$\lim_{s\to 0} sX(s)=\lim_{s\to 0}\int_{0^-}^{\infty} x'(t)e^{-st}\,dt+x(0^-)=x(t)\bigg|_{0^-}^{\infty}+x(0^-).$$

Portanto, o teorema do valor final afirma que

$$x(\infty)=\lim_{s\to 0}sX(s).$$

**Exemplo** Seja o sistema descrito pela função de transferência $H(s)=\frac{s+5}{s^2+4s+3}$ e entrada $x(t)=e^{-2t}u(t)$.

```python
from lcapy import *
H=tf([1,5],[1,4,3])
H
```

```python
v=exp(-2*t)
X=v.LT()
Y=H*X
Y
```

```python
y=Y.ILT(causal=True)
y
```

```python
y.plot((0,10));
```

A resposta inicia com valor nulo,  e depois de uma oscilação transitória retorna à zero. Isto poderia ser visto diretamente com os teoremas do valor inicial e final, sem fazer a transformada inversa. 
$$y(0^+)=\lim_{s\to\infty}sY(s)=\lim_{s\to\infty}\frac{s(s+5)}{(s^2+4s+3)(s+2)}=0$$
e
$$y(\infty)=\lim_{s\to 0}sY(s)=\lim_{s\to 0}\frac{s(s+5)}{(s^2+4s+3)(s+2)}=0$$

**Exemplo** Análise de Laplace de um filtro passa-baixas $RC$

A *netlist* abaixo descreve um filtro passa-baixas $RC$ de primeira ordem. Vamos estudar o comportamento do filtro para 3 frequências, a frequência de corte ($f_c=2\; H_z$), um décimo de $f_c$ e dez vezes $f_c$.

```python
from lcapy import Circuit,t,voltage,sin,u,s

a = Circuit("""
P1 1 0; down=1.5, v_=v_i(t)
R 1 2 2; right=1.5
C 2 0_2 {1/4}; down
W 0 0_2; right
W 2 3; right
W 0_2 0_3; right
P2 3 0_3; down, v^=v_o(t)
;draw_nodes=connections""")
```

```python
a.draw()
```

Onde $v_i(t)$ é a tensão de entrada e $v_0(t)$ é a tensão de saída. A função de transferência do filtro pode ser encontrada especificando os nós, ou seja,

```python
H = a.transfer(1, 0, 3, 0)
H
```

A função de transferência também pode ser obtida especificando as componentes P1 e P2,

```python
H = a.P1.transfer('P2')

H
```

Vamos definir um sinal senoidal de entrada com frequência angular $\omega=3 \frac{\mbox{rad}}{\mbox{s}}$, chaveado em $t=0$.

```python
v_i1 = voltage(sin(0.2* t) * u(t))
v_i2 = voltage(sin(2* t) * u(t))
v_i3 = voltage(sin(20* t) * u(t))
```

A transformada de Laplace da tensão de entrada é obtida como:

```python
V_i1 = v_i1(s)
V_i1
```

```python
V_i2 = v_i2(s)
V_i2
```

```python
V_i3 = v_i3(s)
V_i3
```

A transformada de Laplace da saída é encontrada usando o teorema da convolução:

```python
V_o1 = V_i1 * H
V_o1
```

```python
V_o2 = V_i2 * H
V_o2
```

```python
V_o3 = V_i3 * H
V_o3
```

Esta função tem 3 polos, que podem ficar explícitos com a forrma *zero-pole-gain* da resposta.

```python
V_o1.ZPK()
```

```python
V_o2.ZPK()
```

```python
V_o3.ZPK()
```

A transformada inversa fornece a tensão de saída no domínio do tempo. O resultado pode ser simplificado com o método simplify_terms.

```python
v_o1 = V_o1(t)
v_o1.simplify_terms()
```

```python
v_o2 = V_o2(t)
v_o2.simplify_terms()
```

```python
v_o3 = V_o3(t)
v_o3.simplify_terms()
```

Finalmente, podemos fazer um gráfico dos sinais de entrada e saída. 

```python
ax = v_i1.plot((-1, 40), label='input $w=0.2$')
ax = v_o1.plot((-1, 40), axes=ax, label='output $w=0.2$')
ax.legend();
```

Em baixas frequências a forma de onda na saída é quase uma cópia da tensão de entrada. 

```python
ax = v_i2.plot((-1, 10), label='input $w=2$')
ax = v_o2.plot((-1, 10), axes=ax, label='output $w=2$')
ax.legend();
```

Na frequência de corte, a amplitude da forma de onda é reduzida para $\frac{1}{\sqrt 2}$ da amplitude máxima da tensão de entrada;

```python
ax = v_i3.plot((-1, 5), label='input $w=20$')
ax = v_o3.plot((-1, 5), axes=ax, label='output $w=20$')
ax.legend();
```

Finalmente, para $f=10f_c$ a amplitude da forma de onda de saída é um décimo da amplitude da tensão de entrada. Veja também a influência do polo $s=2$ da função de transferência, principalmente nas duas últimas figuras, devido ao termo $e^{-2t}$. Para a primeira figura a amplitude deste termo é muito pequena (aproximadamente um centésimo), e o transitório é quase inexistente. 

## Estabilidade

Da equação $Q(s)Y(s)=P(s)X(s)$ obtemos $H(s)=\frac{P(s)}{Q(s)}$. Qual a relação entre $Q(s)$ e $Q(\lambda)$? São iguais? $Q(\lambda)$ é o polinômio característico do sistema, e descreve os seus modos característicos. Daí, $Q(\lambda)$ é uma descrição interna do sistema. Já $H(s)$ é uma relação de entrada-saída, não considerando o sistema internamente. $H(s)$ fornece uma descrição externa do sistema. Entretanto, se ao montar a função de transferência, não houver cancelamentos entre polos e zeros iguais no numerador e denominador de $H(s)$, então deste caso, $Q(s)=Q(\lambda)$, e portanto de Q(s) poderemos obter também informação sobre a estabilidade interna do sistema. 

Assim, a princípio, podemos dizer que se todos polos de $H(s)$, ou as raízes de $Q(s)$, são negativos, ou complexos com parte real negativa (os polos estão no semiplano esquerdo do plano complexo), então, o sistema é BIBO estável. Pois $\mathcal{L^{-1}}(\frac{A}{s+a})=Ae^{-at}$, de forma que a resposta ao impulso tende assintoticamente a zero sendo portanto absolutamente integrável. 

Se a função de transfererência não é própria, isso é, $M$, o grau $P(s)$, é maior que $N$, o grau de $Q(s)$, então o sistema é instável. Por exemplo, seja 

$$H(s)=\frac{s^3+4s^2+4s+5}{s^2+3s+2}=s+\frac{s^2+2s+5}{s^2+3s+2}.$$

Note o termo $s$, o primeiro termo de $H(s)$. Se um degrau ($X(s)=\frac{1}{s}$) é aplicado à entrada, a saída conterá um impulso, sendo portanto ilimitada e BIBO instável. 

Quando $Q(s)=Q(\lambda)$, então de $Q(s)$ podemos determinar a estabilidade quase da mesma forma que antes. 

* Se os polos de $H(s)$ estão no semiplano complexo esquerdo, o sistema é assintoticamente estável.
* Se um ou mais polos estão no semi plano direito, ou ainda, existem polos repetidos com parte real nula, o sistema é internamente instável e daí também BIBO instável. Note que $\mathcal{L}^{-1}(\frac{1}{s^2})=tu(t)$, instável.
* Se existem polos não repetidos com parte real nula, e nenhum polo no caso anterior, então o sistema é marginalmente estável e BIBO instável. Mote que $\mathcal{L}^{-1}(\frac{1}{s})=u(t)$.
