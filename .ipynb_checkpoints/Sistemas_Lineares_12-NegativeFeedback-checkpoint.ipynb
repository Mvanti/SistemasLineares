## Sensibilidade de Ganho e Largura de Banda

Um amplificador operacional configuarado como amplificador no inversor com realimentao negativa  mostrado na figura abaixo:






No circuito acima, o ganho de malha aberta em baixas frequncias *A*  aproximadamente 2M, ou 126 dB. O Ganho de realimentao negativa, $\beta$  dado pelo divisor de tenso com os resistores $R_1$ e $R_2$, ou seja, $\beta=\frac{R_1}{R_1+R_2}=0,1$. Assim, o ganho em malha fechada pode ser calculado como abaixo:

$H(s)=\frac{A(s)}{1+A(s)\beta}$

Note que como ser mostrado em seguida, normalmente $A(s)$ varia (diminui) com a frequncia enquanto o ganho de realimentao  constante. Enquanto $A(s)$  grande, o produto $A(s)\beta$  consideravelmente maior que 1, e a expresso acima pode ser aproximada como

$H(s) \approx \frac{A(s)}{A(s)\beta}=\frac{1}{\beta}.$

No exemplo acima, teremos $H(s) \approx 10$, ou 20 dB, portanto uma reduo de 106 dB em relao ao ganho de malha aberta. Entretanto, o fato  importante aqui  que $A(s)$ comea a ser reduzido em frequncias muto baixas, com uma reduo de 20 dB por dcada, aproximadamente, como um filtro passa baixas de primeira ordem. A reduo do ganho de malha aberta  causada primariamente pela presena de capacitncias parasitas que inserem polos na funo de trasferncia, causando um comportamento pouco regular. Um capacitor de compensao ($C_c$, na figura) adiciona um polo dominante, causando a reduo aproximadamente constante de $20 dB$ por dcada. Assim, para frequncias maiores que a frequncia de corte em malha aberta, $A(s)$ pode ser escrita como :

$A(s)=\frac{A}{\frac{\omega}{\omega_c}}=A\frac{\omega_c}{\omega}$

onde $A$  o ganho $A(s)$ em baixas frequncias e $\omega_c$  a frequncia de corte em malha aberta. Na figura abaixo  mostrado o diagrama de Bode para o os ganhos de malha aberta e fechada: 



No diagrama, a linha azul  o ganho em malha aberta enquanto a linha vermelha  o ganho de malha fechada. O diagrama mostra que apesar da reduo do ganho este  constante at $\omega =\omega^{'}\approx  100 KHz$, enquanto $\omega_c \approx 0,5 Hz$. Verifique que $\frac{\omega^{'}}{\omega_c}=200.10^3$, que  a mesma relao na qual o ganho  reduzido, isto , $\frac{2.10^6}{10}=200.10^3$.
Este resultado pode ser confirmado observando que na frequncia $\omega^{'}$ podemos escrever 

$\frac{A}{1+A\beta}=A\frac{\omega_c}{\omega^{'}}\rightarrow \omega^{'}=(1+A\beta)\omega_c.$

Portanto, diminuir o ganho do amplificador por um determinado fator faz com que a largura de banda aumente pelo mesmo fator. Como resultado, um dos parmetros mais importantes em um amplificador operacional  o produto ganho x largura de banda, ou GBP na sigla em ingles, fornecido nas especificaes do componente, e com o qual dada uma largura de banda desejada se encontra o ganho correspondente ou vice-versa.
